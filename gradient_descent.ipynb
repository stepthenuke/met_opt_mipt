```python
import numpy as np
```

```python
class Function:

    def __init__(self, func, grad, hesse):
        self.f = func
        self.g = grad
        self.h = hesse
    
    def calculate(self, x):
        return self.f(x)

    def gradient(x):
        return self.g(x)

    def hesse(x):
        return self.h(x)
```

```python
def grad_descent(f:Function, x, tol, tol_1d, step=None, max_it=1000):

        def l(x):
            return 0.01

        def step(x):
            y = np.zeros(len(x))
            for i in range(len(x)):
                y[i] = x[i] - l(x) * f.gradient(x)[i]
            return y

        def dist(x, y):
            return np.linalg.norm(x-y)

        prev_x = x
        x = step(x)
        it = 0
        while(dist(prev_x, x) > tol and it < max_it):
            it += 1
            prev_x = x
            x = step(x)

        print(it)
        return x
```
